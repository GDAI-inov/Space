```python
def file_list(path):
    return [[os.path.join(path, file) for file in files] for path, _, files in os.walk(path)][0]
```

## 2. Data 전처리

```python
from copy import deepcopy
```

### 2-1. ace

```python
# datetime 객체로 변환

pd.to_datetime(ace[2013]["year"][:5], format="%Y")
```

```python
# day, hour, minute을 더해주기 위해서 timedelta객체로 만듦

ace[2013]["doy"][:5].map(lambda x: pd.Timedelta(x-1, unit='days'))
```

```python
%%time
# 연도가 1월 1일로 설정되어 있으므로 x-1을 함. 

for year, val in ace.items():
    val["dt"] = pd.to_datetime(val['year'], format='%Y') + \
    val["doy"].map(lambda x: pd.Timedelta(x-1, unit='days')) + \
    val["h"].map(lambda x: pd.Timedelta(x, unit='hours')) + \
    val["m"].map(lambda x: pd.Timedelta(x, unit='minutes'))
    
    ace[year] = val.iloc[:, 4:].set_index("dt")
    ace[year].to_csv(f"proc_data/ace/ace_{year}_datetime.csv")
    print(f'{year} is done!')
    
ace[1999].head()
```

### 2-2. target

```python
target.melt(['date']).sort_values(['date']).reset_index(drop=True)
```

```python
target = target.melt(['date'])
target['variable'] = target.variable.map({
                        "kp_0h":0,
                        "kp_3h":3,
                        "kp_6h":6,
                        "kp_9h":9,
                        "kp_12h":12,
                        "kp_15h":15,
                        "kp_18h":18,
                        "kp_21h":21,
                    })
target['dt'] = pd.to_datetime(target["date"]) + \
         target["variable"].apply(pd.Timedelta, unit="h")

target = target.iloc[:, 2:].set_index("dt").sort_index()
target.to_csv("proc_data/target.csv")

target.head()
```

### 2-3. problem

```python
problem["dt"] = pd.to_datetime('2015', format='%Y') + \
problem["doy"].map(lambda x: pd.Timedelta(x-1, unit='days')) + \
problem["h"].map(lambda x: pd.Timedelta(x, unit='hours')) + \
problem["m"].map(lambda x: pd.Timedelta(x, unit='minutes'))
    
problem = problem.iloc[:, 3:].set_index("dt")
problem.to_csv("proc_data/problem.csv")

problem.head()
```
