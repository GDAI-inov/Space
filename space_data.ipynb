### 2-1. ace

```python
# datetime 객체로 변환

pd.to_datetime(ace[2013]["year"][:5], format="%Y")
```

```python
# day, hour, minute을 더해주기 위해서 timedelta객체로 만듦

ace[2013]["doy"][:5].map(lambda x: pd.Timedelta(x-1, unit='days'))
```

```python
%%time
# 연도가 1월 1일로 설정되어 있으므로 x-1을 함. 

for year, val in ace.items():
    val["dt"] = pd.to_datetime(val['year'], format='%Y') + \
    val["doy"].map(lambda x: pd.Timedelta(x-1, unit='days')) + \
    val["h"].map(lambda x: pd.Timedelta(x, unit='hours')) + \
    val["m"].map(lambda x: pd.Timedelta(x, unit='minutes'))
    
    ace[year] = val.iloc[:, 4:].set_index("dt")
    ace[year].to_csv(f"proc_data/ace/ace_{year}_datetime.csv")
    print(f'{year} is done!')
    
ace[1999].head()
```

```python
total = dict()
for year, val in ace.items():
    ins = target.set_index("date").loc[f"{year}-01-01":f"{year}-12-31"]
    ins.reset_index(inplace=True)
    ins = ins.melt(['date']).rename({'variable':'h', 'value':'target'}, axis=1)
    ins['h'] = ins.h.map({f'kp_{str(n)}h': n for n in list(range(0, 24, 3))})
    ins = ins.sort_values(['date', 'h']).reset_index(drop=True)
    
    val_p = val.iloc[:, 4:]
    val_p[val_p <= -9999] = np.nan
    val_p.fillna(method='ffill', inplace=True)
    val_na = val_p.isna().sum().sum()
    while val_na != 0:
        val_p.fillna(val_p.std(), inplace=True)
        val_na = val_p.isna().sum().sum()
    val[val.columns[4:]] = val_p

    val['date'] = pd.to_datetime(val['year'], format='%Y') + \
    val['doy'].map(lambda x: pd.Timedelta(x-1, unit='days'))
    val['h'] = val.h.map(dict(enumerate([n for _ in list(range(0, 24, 3)) for n in itertools.repeat(_, 3)])))
    
    val = val[['h']+list(val.columns[4:])]
    val = val.pivot_table(index=['date', 'h'], aggfunc='mean')
    val.reset_index(inplace=True)
    val['date'] = val.date + val.h.apply(pd.Timedelta, unit="h")
    val = val[['date']+list(val.columns[-3:])+list(val.columns[2:-3])]
    
    total[year] = pd.concat([val, ins.target], axis=1).set_index('date')
    total[year].to_csv(f"proc_data/total/total_{year}.csv")
    print(f'{year} is done!')
    
total[1999].head()
```

```python
problem_p = problem.iloc[:, 3:]
problem_p[problem_p <= -9999] = np.nan
problem_p.fillna(method='ffill', inplace=True)
problem[problem.columns[3:]] = problem_p

problem['date'] = pd.to_datetime(problem['doy'].map(lambda x: x-1), unit='D')
problem['h'] = problem.h.map(dict(enumerate([n for _ in list(range(0, 24, 3)) for n in itertools.repeat(_, 3)])))
    
problem = problem[['h']+list(problem.columns[3:])]
problem = problem.pivot_table(index=['date', 'h'], aggfunc='mean')
problem.reset_index(inplace=True)
problem['date'] = (problem.date + problem.h.apply(pd.Timedelta, unit="h")).dt.strftime('%m-%d %H:%M')
problem = problem[['date']+list(problem.columns[-3:])+list(problem.columns[2:-3])].set_index('date')

problem.to_csv("proc_data/problem.csv")

problem.head()
```
